```python
import pandas as pd
import numpy as np
from tqdm import tqdm



train_df = pd.read_parquet('train_features.parquet')
Start_df = pd.read_csv('Start_df.csv')
Turn_df = pd.read_csv('Turn_df.csv')
Walk_df = pd.read_csv('Walk_df.csv')
```

```python
from sklearn.model_selection import GroupKFold, train_test_split
import pickle

x_train, x_test, y_train, y_test = train_test_split(train_df, Start_df['StartHesitation'], test_size=0.5, random_state= 42, stratify= Start_df['StartHesitation'])

x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
x_train = x_train.to_numpy()
x_test = x_test.to_numpy()
```

```python
x_train = x_train.reshape(x_train.shape[0], x_train.shape[1], 1)
x_test = x_test.reshape(x_test.shape[0], x_test.shape[1], 1)
x_train.shape, x_test.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
def transformer_encoder(inputs, head_size, num_heads, ff_dim, dropout=0):
    # Normalization and Attention
    x = layers.LayerNormalization(epsilon=1e-6)(inputs)
    x = layers.MultiHeadAttention(
        key_dim=head_size, num_heads=num_heads, dropout=dropout
    )(x, x)
    x = layers.Dropout(dropout)(x)
    res = x + inputs

    # Feed Forward Part
    x = layers.LayerNormalization(epsilon=1e-6)(res)
    x = layers.Conv1D(filters=ff_dim, kernel_size=1, activation="relu")(x)
    x = layers.Dropout(dropout)(x)
    x = layers.Conv1D(filters=inputs.shape[-1], kernel_size=1)(x)
    return x + res
```

```python
def build_model(
    input_shape,
    head_size,
    num_heads,
    ff_dim,
    num_transformer_blocks,
    mlp_units,
    dropout=0,
    mlp_dropout=0,
):
    inputs = keras.Input(shape=input_shape)
    x = inputs
    for _ in range(num_transformer_blocks):
        x = transformer_encoder(x, head_size, num_heads, ff_dim, dropout)

    x = layers.GlobalAveragePooling1D(data_format="channels_first")(x)
    for dim in mlp_units:
        x = layers.Dense(dim, activation="relu")(x)
        x = layers.Dropout(mlp_dropout)(x)
    outputs = layers.Dense(1, activation="sigmoid")(x)
    return keras.Model(inputs, outputs)
```

```python
from sklearn.utils import class_weight

weights = class_weight.compute_class_weight(class_weight = 'balanced',
                                            classes = np.unique(y_train),
                                            y = y_train)
```

```python
weights
```

```python
input_shape = x_train.shape[1:]

model = build_model(
    input_shape,
    head_size=256,
    num_heads=4,
    ff_dim=4,
    num_transformer_blocks=4,
    mlp_units=[128],
    mlp_dropout=0.4,
    dropout=0.25,
)

model.compile(
    loss= keras.losses.BinaryCrossentropy(),
    optimizer=keras.optimizers.Adam(),
    metrics=[[keras.metrics.Precision(), keras.metrics.Accuracy(), keras.metrics.AUC(curve= 'PR')]],
)
model.summary()

callbacks = [keras.callbacks.EarlyStopping(patience=10, restore_best_weights=True)]

model.fit(
    x_train,
    y_train,
    validation_data=(x_test, y_test),
    epochs=200,
    batch_size=64,
    callbacks=callbacks,
    class_weight={i:weights[i] for i in range(len(weights))},
)

model.evaluate(x_test, y_test, verbose=1)
```
